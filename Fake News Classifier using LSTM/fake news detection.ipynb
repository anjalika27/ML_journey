Applying one hot encoding after stemming

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
corpus=[]
for i in X:
    content=re.sub('[^a-zA-Z]',' ',i)
    content=content.lower()
    content=content.split()
    review=[ps.stem(word) for word in content if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
corpus
```

```python
from tensorflow.keras.preprocessing.text import one_hot
vocab_size=5000
one_hot_rep=[one_hot(words,vocab_size) for words in corpus]
one_hot_rep
#jis index pe word hoga in vocab uska index return krega 
```

```python
corpus[1],one_hot_rep[1]
```

```python
m=0
for i in one_hot_rep:
   if m<len(i):
    m=len(i)
    
print(m)
#max length of a sequence
```

```python
sent_len=35
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_sent=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_len)
padded_sent
```

```python
#creating model 
embedding_features=40

import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Embedding,LSTM,Dense
```

```python
model=Sequential()
model.add(Embedding(vocab_size,embedding_features,input_length=sent_len))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile('adam','binary_crossentropy','accuracy')
```

```python
model.summary()
```

```python
# model.fit(X,y,epochs=10)
X=np.array(padded_sent)
y=np.array(y)
```

```python
model.fit(X,y,epochs=10)
```

```python
X_test=test['content']
```

```python
ps=PorterStemmer()
corpus_test=[]
for i in X_test:
    content=re.sub('[^a-zA-Z]',' ',i)
    content=content.lower()
    content=content.split()
    review=[ps.stem(word) for word in content if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus_test.append(review)
corpus_test
```

```python
vocab_size=5000
one_hot_rep_test=[one_hot(words,vocab_size) for words in corpus_test]
one_hot_rep_test
```

```python
padded_sent_test=pad_sequences(one_hot_rep_test,padding='pre',maxlen=sent_len)
padded_sent_test
```

```python
X_test=np.array(padded_sent_test)
X_test
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
y_hat=np.where(y_pred>0.6,1,0)
```
